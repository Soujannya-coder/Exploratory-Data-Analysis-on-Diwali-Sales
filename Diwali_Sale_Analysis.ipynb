```python
#Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Loading CSV as Data Frame 
df=pd.read_csv("Diwali Sales Data.csv",encoding='unicode_escape')
```

```python
#Chacking no of rows and columns of the DataFrame
df.shape
```

```python
#Checking First five rows
df.head()
```

```python
#Checking information of the DataFrame
df.info()
```

```python
#Droping status and unnamed1 as its unrelated and blank
df.drop(['Status','unnamed1'],axis=1,inplace=True)#inplace =True saves current changes in dataframe 
```

```python
#Counting no of null values
pd.isnull(df).sum()
```

```python
#droping null values
df.dropna(inplace=True)
```

```python
#Counting no of null values after removing nulls
pd.isnull(df).sum()
```

```python
#Changing Data Type
df['Amount']=df['Amount'].astype('int64')
```

```python
#chacking data type
df['Amount'].dtypes
```

```python
#Checking list of columns
df.columns
```

```python
#rename columns
df.rename(columns={'Marital_Status':'Married'})
```

```python
#checking descripton of data frame
df.describe()
```

```python
#checking descripton of specific columns in data frame
df[['Age','Orders','Amount']].describe()
```

<h1>Exploratory Data Analysis</h1>


```python
#Gender
ax=sns.countplot(x='Gender',data=df)
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

From above graph we can see female byuers spend approximately double amount on shopping

```python
#Age
ax=sns.countplot(x='Age Group',hue='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

From the above graph we can see most of the byers are of age group between 26-35

```python
#State
ax=sns.countplot(data=df,x='State',hue='Gender')
sns.set(rc={'figure.figsize':(50,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total no of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

From the above graph we can see maximum no of byers are from uttar Pradesh

```python
#total amount/sales from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

```python
#Marital status
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(7,5)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

From the above graphs we can see most of the byers are married(women) and they have high purchasing power

```python
#occupation
ax=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(22,5)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_state=df.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=sales_state,x='Occupation',y='Amount',hue='Gender')
```

From above graph we can see the most of the byers are working in IT,Aviation and Healthcare sector

```python
#Product Category
ax=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(22,10)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_state=df.groupby(['Product_Category','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(35,10)})
sns.barplot(data=sales_state,x='Product_Category',y='Amount',hue='Gender')
```

From above graph we can see that most of the sold products are from food ,footwear &elctronics
